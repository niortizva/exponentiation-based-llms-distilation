```python
import pandas as pd
```

```python
df = pd.read_csv("computation_efficiency.csv")
df = df.sort_values(by=["model", "vocab_size"])

df_cols = ["per_batch_mean_s", "per_batch_variance_s", "per_token_mean_s", "per_token_variance_s"]
df_us_cols = ["per_batch_mean_us", "per_batch_variance_us", "per_token_mean_us", "per_token_variance_us"]

df = df.rename(columns={col: new_col for col, new_col in zip(df_cols, df_us_cols)})
for new_col in df_us_cols:
    df[new_col] = df[new_col] * 1e6  # seconds to microseconds
df = df.sort_values(by=["vocab_size"])
df
```

```python
for (
    vocab_s, model_name, batches, per_batch, per_batch_v, 
    per_token, per_token_v
 ) in df.values:
    print(f"{model_name} & {vocab_s} & {batches} &" +
          f" {per_batch:.4f} $\\pm$ {per_batch_v:.4f} &" + 
          f" {per_token:.4f} $\\pm$ {per_token_v:.4f} \\\\")
```
